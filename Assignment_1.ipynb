```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("/content/3.jpeg")
plt.axis("off")
plt.imshow(image)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
cv2.imwrite("gray_image.jpg", gray_image)
```

```python
bright_image = cv2.add(gray_image, 50)
```

```python
dark_image = cv2.subtract(gray_image, 50)
```

```python
_, binary_image = cv2.threshold(gray_image, 128, 255, cv2.THRESH_BINARY)
```

```python
negative_image = 255 - gray_image
```

```python
log_image = np.log1p(gray_image.astype(np.float32))
log_image = (log_image / np.max(log_image) * 255).astype(np.uint8)
```

```python
gamma = 0.5
inv_gamma = 1.0 / gamma
table = np.array([((i / 255) ** inv_gamma) * 255 for i in range(256)]).astype("uint8")
gamma_image = cv2.LUT(gray_image, table)
```

```python
images = [gray_image, bright_image, dark_image, binary_image,
          negative_image, log_image, gamma_image]

titles = ['Grayscale', 'Bright +50', 'Dark -50', 'Threshold',
          'Negative', 'Log Transform', 'Gamma Correction']

plt.figure(figsize=(15, 10))

for i in range(len(images)):
    plt.subplot(2, 4, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
