```python
import random
import cv2

def add_noise(img):
    # Getting the dimensions of the image
    row, col = img.shape

    number_of_pixels = random.randint(300, 10000)
    for i in range(number_of_pixels):
        y_coord = random.randint(0, row - 1)
        x_coord = random.randint(0, col - 1)
        img[y_coord, x_coord] = 255

    number_of_pixels = random.randint(300, 10000)
    for i in range(number_of_pixels):
        y_coord = random.randint(0, row - 1)
        x_coord = random.randint(0, col - 1)
        img[y_coord, x_coord] = 0

    return img

img = cv2.imread('/content/camera.jpeg', cv2.IMREAD_GRAYSCALE)

sp_noisy_img = add_noise(img)

cv2.imwrite('salt-and-pepper-camera.jpg', sp_noisy_img)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("Salt & Pepper Noisy Image")
plt.imshow(sp_noisy_img, cmap='gray')
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Histogram")
plt.hist(sp_noisy_img.ravel(), bins=256, range=[0,256], color='black')
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to add Gaussian noise
def add_gaussian_noise(img, mean=0, var=30):
    row, col = img.shape
    sigma = var ** 0.5
    gauss = np.random.normal(mean, sigma, (row, col))
    noisy = img + gauss
    noisy = np.clip(noisy, 0, 255).astype(np.uint8)
    return noisy

img = cv2.imread('/content/camera.jpeg', cv2.IMREAD_GRAYSCALE)

gaussian_noisy_img = add_gaussian_noise(img)

cv2.imwrite("gaussian-noise-camera.jpg", gaussian_noisy_img)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("Gaussian Noisy Image")
plt.imshow(gaussian_noisy_img, cmap='gray')
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Histogram")
plt.hist(gaussian_noisy_img.ravel(), bins=256, range=[0,256], color='black')
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random

# Function to add Impulse Noise
def add_impulse_noise(img, prob=0.02):
    row, col = img.shape
    noisy = img.copy()

    num_pixels = int(prob * row * col)

    for i in range(num_pixels):
        y_coord = random.randint(0, row - 1)
        x_coord = random.randint(0, col - 1)
        noisy[y_coord, x_coord] = random.randint(0, 255)

    return noisy

img = cv2.imread('/content/camera.jpeg', cv2.IMREAD_GRAYSCALE)

impulse_noisy_img = add_impulse_noise(img, prob=0.02)

cv2.imwrite("impulse-noise-camera.jpg", impulse_noisy_img)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("Impulse Noisy Image")
plt.imshow(impulse_noisy_img, cmap='gray')
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Histogram")
plt.hist(impulse_noisy_img.ravel(), bins=256, range=[0,256], color='black')
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.show()

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to add Speckle noise
def add_speckle_noise(img):
    row, col = img.shape
    # Generate Gaussian noise
    gauss = np.random.randn(row, col)
    noisy = img + img * gauss * 0.2

    noisy = np.clip(noisy, 0, 255).astype(np.uint8)
    return noisy

img = cv2.imread('/content/camera.jpeg', cv2.IMREAD_GRAYSCALE)

speckle_noisy_img = add_speckle_noise(img)

cv2.imwrite("speckle-noise-camera.jpg", speckle_noisy_img)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("Speckle Noisy Image")
plt.imshow(speckle_noisy_img, cmap='gray')
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Histogram")
plt.hist(speckle_noisy_img.ravel(), bins=256, range=[0,256], color='black')
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.show()


```
