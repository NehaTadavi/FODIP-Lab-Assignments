```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def show_image_with_hist(title, img):
    plt.figure(figsize=(10,4))

    plt.subplot(1,2,1)
    plt.title(title)
    plt.imshow(img, cmap='gray')
    plt.axis("off")

    plt.subplot(1,2,2)
    plt.title(title + " Histogram")
    plt.hist(img.ravel(), bins=256, range=[0,256], color='black')
    plt.xlabel("Pixel Intensity")
    plt.ylabel("Frequency")

    plt.tight_layout()
    plt.show()


img = cv2.imread("/content/low .jpeg", cv2.IMREAD_GRAYSCALE)

low_contrast_img = cv2.normalize(img, None, alpha=100, beta=150, norm_type=cv2.NORM_MINMAX)

r_min, r_max = np.min(low_contrast_img), np.max(low_contrast_img)
contrast_stretched = ((low_contrast_img - r_min) / (r_max - r_min) * 255).astype(np.uint8)

hist_equalized = cv2.equalizeHist(low_contrast_img)

show_image_with_hist("Original Low Contrast Image", low_contrast_img)
show_image_with_hist("Contrast Stretched Image", contrast_stretched)
show_image_with_hist("Histogram Equalized Image", hist_equalized)
```
